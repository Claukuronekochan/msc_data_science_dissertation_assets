```python
consolidated = countries_all.melt(id_vars=['Country Name', 'Region', 'Sub-region', 'Country Code', 'Indicator Name'], 
        var_name="Year", 
        value_name="Value")
consolidated
```

```python
consolidated.loc[consolidated['Country Code'] == 'WLD', ['Country Name','Region','Sub-region']] = 'World'
consolidated['Group']=consolidated['Region']
consolidated.loc[consolidated['Country Name'] == 'Bangladesh', ['Group']] = 'Bangladesh'
consolidated
```

```python
consolidated['Value'] = consolidated.groupby(['Country Name','Indicator Name'])['Value'].ffill()
consolidated['Value'] = consolidated.groupby(['Country Name','Indicator Name'])['Value'].bfill()
consolidated
```

```python
consolidated.to_csv('historical_world_data.csv')
consolidated['Group'].unique()
```

```python
#world population
p_world_i = consolidated[consolidated['Indicator Name'] == 'Population, total']
p_asia_i = p_world_i[p_world_i['Region'] == 'Asia']
p_bangladesh_i = p_asia_i[p_asia_i['Country Name'] == 'Bangladesh']

#energy access
a_world_i = consolidated[consolidated['Indicator Name'] == 'Access to electricity (% of population)']
a_asia_i = a_world_i[a_world_i['Region'] == 'Asia']
a_bangladesh_i = a_asia_i[a_asia_i['Country Name'] == 'Bangladesh']
r_access = consolidated[consolidated['Indicator Name'] == 'Access to electricity, rural (% of rural population)']
a_chars_i = r_access[r_access['Country Name'] == 'Bangladesh']

#power consumption
pc_world_i = consolidated[consolidated['Indicator Name'] == 'Electric power consumption (kWh per capita)']
pc_asia_i = pc_world_i[pc_world_i['Region'] == 'Asia']
pc_bangladesh_i = pc_asia_i[pc_asia_i['Country Name'] == 'Bangladesh']

#energy sources
es_world_coal = consolidated[consolidated['Indicator Name'] == 'Electricity production from coal sources (% of total)']
es_asia_coal = es_world_coal[es_world_coal['Region'] == 'Asia']
es_bangladesh_coal = es_asia_coal[es_asia_coal['Country Name'] == 'Bangladesh']

es_world_hydro = consolidated[consolidated['Indicator Name'] == 'Electricity production from hydroelectric sources (% of total)']
es_asia_hydro = es_world_hydro[es_world_hydro['Region'] == 'Asia']
es_bangladesh_hydro = es_asia_hydro[es_asia_hydro['Country Name'] == 'Bangladesh']

es_world_gas = consolidated[consolidated['Indicator Name'] == 'Electricity production from natural gas sources (% of total)']
es_asia_gas = es_world_gas[es_world_gas['Region'] == 'Asia']
es_bangladesh_gas = es_asia_gas[es_asia_gas['Country Name'] == 'Bangladesh']

es_world_nuclear = consolidated[consolidated['Indicator Name'] == 'Electricity production from nuclear sources (% of total)']
es_asia_nuclear = es_world_nuclear[es_world_nuclear['Region'] == 'Asia']
es_bangladesh_nuclear = es_asia_nuclear[es_asia_nuclear['Country Name'] == 'Bangladesh']

es_world_oil = consolidated[consolidated['Indicator Name'] == 'Electricity production from oil sources (% of total)']
es_asia_oil = es_world_oil[es_world_oil['Region'] == 'Asia']
es_bangladesh_oil = es_asia_oil[es_asia_oil['Country Name'] == 'Bangladesh']

es_world_renewable = consolidated[consolidated['Indicator Name'] == 'Electricity production from renewable sources, excluding hydroelectric (% of total)']
es_asia_renewable = es_world_renewable[es_world_renewable['Region'] == 'Asia']
es_bangladesh_renewable = es_asia_renewable[es_asia_renewable['Country Name'] == 'Bangladesh']

#extreme weather
ex_world_i = consolidated[consolidated['Indicator Name'] == 'Droughts, floods, extreme temperatures (% of population, average 1990-2009)']
ex_asia_i = ex_world_i[ex_world_i['Region'] == 'Asia']
ex_bangladesh_i = ex_asia_i[ex_asia_i['Country Name'] == 'Bangladesh']

#internet usage
i_world_i = consolidated[consolidated['Indicator Name'] == 'Individuals using the Internet (% of population)']
i_asia_i = i_world_i[i_world_i['Region'] == 'Asia']
i_bangladesh_i = i_asia_i[i_asia_i['Country Name'] == 'Bangladesh']

#mortality
m_world_pollution = consolidated[consolidated['Indicator Name'] == 'Mortality rate attributed to household and ambient air pollution, age-standardized (per 100,000 population)']
m_asia_pollution = m_world_pollution[m_world_pollution['Region'] == 'Asia']
m_bangladesh_pollution = m_asia_pollution[m_asia_pollution['Country Name'] == 'Bangladesh']

m_world_water = consolidated[consolidated['Indicator Name'] == 'Mortality rate attributed to unsafe water, unsafe sanitation and lack of hygiene (per 100,000 population)']
m_asia_water = m_world_water[m_world_water['Region'] == 'Asia']
m_bangladesh_water = m_asia_water[m_asia_water['Country Name'] == 'Bangladesh']

#water access
w_world_i = consolidated[consolidated['Indicator Name'] == 'People using safely managed drinking water services (% of population)']
w_asia_i = w_world_i[w_world_i['Region'] == 'Asia']
w_bangladesh_i = w_asia_i[w_asia_i['Country Name'] == 'Bangladesh']
r_water = consolidated[consolidated['Indicator Name'] == 'People using safely managed drinking water services, rural (% of rural population)']
w_chars_i = r_water[r_water['Country Name'] == 'Bangladesh']
```

```python
world_total_population = p_world_i['Value'].sum()
asia_total_population = p_asia_i['Value'].sum()
bangladesh_total_population = p_bangladesh_i['Value'].sum()

avg_world_energy_access = a_world_i['Value'].mean()
avg_asia_energy_access = a_asia_i['Value'].mean()
bangladesh_energy_access = a_bangladesh_i['Value'].sum()
chars_energy_access = a_chars_i['Value'].sum()

avg_world_power_consumption = pc_world_i['Value'].mean()
avg_asia_power_consumption = pc_asia_i['Value'].mean()
bangladesh_power_consumption = pc_bangladesh_i['Value'].sum()

#energy sources
world_coal = es_world_coal['Value'].mean()
asia_coal = es_asia_coal['Value'].mean()
bangladesh_coal = es_bangladesh_coal['Value'].mean()

world_hydro = es_world_hydro['Value'].mean()
asia_hydro = es_asia_hydro['Value'].mean()
bangladesh_hydro = es_bangladesh_hydro['Value'].mean()

world_gas = es_world_gas['Value'].mean()
asia_gas = es_asia_gas['Value'].mean()
bangladesh_gas = es_bangladesh_gas['Value'].mean()

world_nuclear = es_world_nuclear['Value'].mean()
asia_nuclear = es_asia_nuclear['Value'].mean()
bangladesh_nuclear = es_bangladesh_nuclear['Value'].mean()

world_oil = es_world_oil['Value'].mean()
asia_oil = es_asia_oil['Value'].mean()
bangladesh_oil = es_bangladesh_oil['Value'].mean()

world_renewable = es_world_renewable['Value'].mean()
asia_renewable = es_asia_renewable['Value'].mean()
bangladesh_renewable = es_bangladesh_renewable['Value'].mean()

#extreme weather
extreme_world = ex_world_i['Value'].mean()
extreme_asia = ex_asia_i['Value'].mean()
extreme_bangladesh = ex_bangladesh_i['Value'].mean()

#internet usage
internet_world = i_world_i['Value'].mean()
internet_asia = i_asia_i['Value'].mean()
internet_bangladesh = i_bangladesh_i['Value'].mean()

#mortality
mortality_pollution_world = m_world_pollution['Value'].mean()/100000
mortality_pollution_asia = m_asia_pollution['Value'].mean()/100000
mortality_pollution_bangladesh = m_bangladesh_pollution['Value'].mean()/100000

mortality_world_water = m_world_water['Value'].mean()/100000
mortality_asia_water = m_asia_water['Value'].mean()/100000
mortality_bangladesh_water = m_bangladesh_water['Value'].mean()/100000

#water access
world_water_access = w_world_i['Value'].mean()
asia_water_access = w_asia_i['Value'].mean()
bangladesh_water_access = w_bangladesh_i['Value'].sum()
chars_water_access = w_chars_i['Value'].sum()
```

```python
#summary dataset
data = np.array([['Indicator','Section','Value'],
                 #population
                 ['Population','World',world_total_population],
                 ['Population','Asia',asia_total_population],
                 ['Population','Bangladesh',bangladesh_total_population],
                 ['Population','Chars',total_chars],
                 #electricity access
                 ['Access to electricity','World',avg_world_energy_access],
                 ['Access to electricity','Asia',avg_asia_energy_access],
                 ['Access to electricity','Bangladesh',bangladesh_energy_access],
                 ['Access to electricity','Bangladesh',chars_energy_access],
                 #electricity consumption
                 ['Electric power consumption','World',avg_world_power_consumption],
                 ['Electric power consumption','Asia',avg_asia_power_consumption],
                 ['Electric power consumption','Bangladesh',bangladesh_power_consumption],
                 #energy sources
                 ['Coal','World',world_coal],
                 ['Coal','Asia',asia_coal],
                 ['Coal','Bangladesh',bangladesh_coal],
                 
                 ['Hydro','World',world_hydro],
                 ['Hydro','Asia',asia_hydro],
                 ['Hydro','Bangladesh',bangladesh_hydro],
                 
                 ['Gas','World',world_gas],
                 ['Gas','Asia',asia_gas],
                 ['Gas','Bangladesh',bangladesh_gas],
                 
                 ['Nuclear','World',world_nuclear],
                 ['Nuclear','Asia',asia_nuclear],
                 ['Nuclear','Bangladesh',bangladesh_nuclear],
                 
                 ['Oil','World',world_oil],
                 ['Oil','Asia',asia_oil],
                 ['Oil','Bangladesh',bangladesh_oil],
                 
                 ['Renewable','World',world_renewable],
                 ['Renewable','Asia',asia_renewable],
                 ['Renewable','Bangladesh',bangladesh_renewable],
                 #extreme weather
                 ['Population affected by extreme weather','World',extreme_world],
                 ['Population affected by extreme weather','Asia',extreme_asia],
                 ['Population affected by extreme weather','Bangladesh',extreme_bangladesh],
                 #internet usage
                 ['Internet usage','World',internet_world],
                 ['Internet usage','Asia',internet_asia],
                 ['Internet usage','Bangladesh',internet_bangladesh],
                 #mortality
                 ['Pollution Mortality','World',mortality_pollution_world],
                 ['Pollution Mortality','Asia',mortality_pollution_asia],
                 ['Pollution Mortality','Bangladesh',mortality_pollution_bangladesh],
                 
                 ['Unsafe water mortality','World',mortality_world_water],
                 ['Unsafe water mortality','Asia',mortality_asia_water],
                 ['Unsafe water mortality','Bangladesh',mortality_bangladesh_water],
                 #water access
                 ['Water access','World',world_water_access],
                 ['Water access','Asia',asia_water_access],
                 ['Water access','Bangladesh',bangladesh_water_access],
                 ['Water access','Chars',chars_water_access]])
```
