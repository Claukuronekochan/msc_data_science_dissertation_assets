```python
import pandas as pd
import numpy as np
import dateutil.parser
import datetime
```

```python
charge_data = pd.read_csv("tcp_pv_sample_data_12-2018-01-2019.csv")
```

```python
#charge_data.shape
charge_data.columns
```

```python
charge_data = charge_data.pivot(index='Time', columns='Series', values='Value')
#charge_data = charge_data.fillna(0)
```

```python
#charge_data.isna().sum()
charge_data.reset_index(level=0, inplace=True)
```

```python
charge_data['TimeT'] = [dateutil.parser.parse(d) for d in charge_data['Time']]
```

```python
charge_data['Date'] = [d.date() for d in charge_data['TimeT']]
charge_data['Time'] = [d.time() for d in charge_data['TimeT']]
```

```python
currentDT = datetime.datetime.now()
currentDT = currentDT.strftime("%Y-%m-%d %H:%M:%S+00:00")
currentDT
```

```python
max_date = max(charge_data['TimeT'])

new_dates = pd.date_range(start = max_date, end = currentDT, freq='1H')#periods=5
N_Date = [d.date() for d in new_dates]
N_Time = [d.time() for d in new_dates]
#new_dates
```

```python
df1 = df = pd.DataFrame({'TimeT':new_dates,
                         'Date':N_Date,
                         'Time':N_Time}) 
```

```python
charge_data = charge_data.append(df1, ignore_index = True)
charge_data.shape
```

```python
cols = ['Battery Current', 'Battery Power In', 'Battery Power Out',
        'Battery Voltage', 'Humidity (Sensor)', 'Irradiance', 'Panel Current',
        'Temperature', 'USB Hub Current']
charge_data[cols] = charge_data[cols].replace({'0':np.nan, 0:np.nan})
```

```python
cols2 = ['Time','Battery Current', 'Battery Power In', 'Battery Power Out',
        'Battery Voltage', 'Humidity (Sensor)', 'Irradiance', 'Panel Current',
        'Temperature', 'USB Hub Current']
charge_data[cols2] = charge_data.groupby(['Time'])[cols].ffill()
charge_data[cols2] = charge_data.groupby(['Time'])[cols].bfill()
charge_data[cols] = charge_data[cols].ffill()
```

```python
charge_data#.isna().sum()
#charge_data.groupby(['Time'])[cols].ffill()
#charge_data.groupby(['Time'])[cols].bfill()
```

```python
charge_data = charge_data[['TimeT', 'Date', 'Time', 'Battery Current', 'Battery Power In', 'Battery Power Out',
       'Battery Voltage', 'Humidity (Sensor)', 'Irradiance', 'Panel Current',
       'Temperature', 'USB Hub Current']]
charge_data
```

```python
#random powerbanks distributed every three hours between 100-150 daily
charge_data['Powerbanks Distributed'] = np.random.randint(13, 19, charge_data.shape[0])
#random income per three hours
charge_data['Income'] = charge_data['Powerbanks Distributed'].multiply(np.random.randint(5, 10)/100)
#lightbulb light 192w = 1hour of use of the powerbank with a 3w lightbulb  
charge_data['lightbulbs'] = charge_data['Powerbanks Distributed'].multiply(12)
#powerbank of 16Ah 12V using a fluorescent lightbulb of 15w 
charge_data['energy Cost Production'] = charge_data['Income'].multiply(0.2)
charge_data['Revenue'] = charge_data['Income'].multiply(0.8)
charge_data
```

```python
charge_data.to_csv('charge_data.csv')
```
